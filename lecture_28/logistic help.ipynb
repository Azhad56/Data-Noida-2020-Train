```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-10, 10, 100)
```

```python
y = np.sin(X) + 3
```

```python
plt.plot(X, y)
```

```python
plt.plot(X, y)

plt.plot(X, np.log(y))
```

```python
X = np.linspace(-1000, 1000, 1000)
```

```python
y = 1 / (1 + np.exp(-X))
```

```python
plt.plot(X, y)
```
